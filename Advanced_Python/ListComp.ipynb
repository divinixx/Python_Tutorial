```python
lst1=[]
def lst_sq(lst):
    for i in lst:
        lst1.append(i*i)
    return lst1
```

```python
lst_sq([1,2,3,4,5,6,7])
```

```python
## String Formatting in Python
def greeting(name):
    return "Hello {}. Welcome to the community".format(name)
```

```python
greeting("krish")
```

```python
def welcome_email(name, age):
    return "Welcome {name}. Your age is {age}".format(name=name, age=age)
```

```python
welcome_email('Divyanshu', 20)
```

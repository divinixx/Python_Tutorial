```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

Correlation with HeatMap
# A correlation heatmap uses colored cells typically in a monochromatic scale, to show a 2D correlation
# matrix between two discrete dimensions or event types. It is very important in Feature Selection

```python
df.dtypes
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
## Join Plot
## Unvariate Analysis

sns.jointplot(x='tip', y='total_bill', data=df, kind='hex' )
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg' )
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
df['smoker'].value_counts()
```

```python
sns.displot(df['tip'], kde=False,  bins=10 )
```

```python
df['sex'].value_counts()
```

```python
## COUNT PLOT

sns.countplot(y="sex", data=df)
```

```python
sns.countplot(x="sex", data=df)
```

```python
## Bar Plot
sns.barplot(x='total_bill', y='sex', data=df)
```

```python
# Box Plot 

sns.boxplot(x='sex', y='total_bill', data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df, palette='rainbow',  hue="total_bill", legend=False)
```

```python
sns.boxplot(data=df, orient='v')
```

```python
sns.boxplot(x="total_bill", y="day", hue="smoker", data=df)
```

```python
sns.violinplot(x="total_bill", y="day", data=df, palette="rainbow" , hue="day", legend=False)
```
